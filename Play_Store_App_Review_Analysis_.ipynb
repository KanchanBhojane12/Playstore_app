**In  Install column there is '+' and ',' which need to be filtered than convert to numeric**

````python
# This function is omit the + and , in the data and turn into numeric

def remove_comma(val):
  if '+' and ',' in val:
    data = int(val[:-1].replace(',',''))
    return data
  elif '+' in val:
    data = int(val[:-1])
    return data
  elif ',' in val:
    data = int(val.replace(',',''))
    return data
  else:
    return int(val)
````

**Datatype of Size Column**

````python
# format of the data of size column
data['Size'].unique()
````

Here M denontes the size in MB(MegaByte) and k denotes as kB(kiloByte).Before converting the data, unit of the size should be unique and 'k' and 'M' should be omitted.

````python
# create a function to convert the data into numerical and unit of MB
def size_convert(app_size):
  '''
  This is the function to convert the size into kb to mb
  '''
  try:
    if app_size[-1]=='M':
      return float(app_size[:-1])
    elif app_size[-1]=='k':
      return round((float(app_size[:-1])/1024),2)
    else:
      return float(app_size)
  except:
    return app_size
````

````python
#convert the column as suggested
data['Size']=data['Size'].apply(lambda size:size_convert(size))
#check the datatype of the column after applying function on the column
data['Size'].dtype
````

## Merging the two dataframe

````python
# It creates a summary table for each app showing the count of positive, neutral, and negative sentiments, along with the average sentiment polarity and subjectivity.

# 1. Pivot to get sentiment counts per app
sentiment_counts = df.pivot_table(
    index='App',
    columns='Sentiment',
    aggfunc='size',
    fill_value=0            # Replace all missing values (NaN) with 0 in the pivot table.
).reset_index()

sentiment_counts = sentiment_counts[['App', 'Positive', 'Neutral', 'Negative']]

# 2. Calculate mean polarity and subjectivity per app
mean_sentiments = df.groupby('App')[['Sentiment_Polarity', 'Sentiment_Subjectivity']].mean().reset_index()

# 3. Merge counts with mean values
final_df = pd.merge(sentiment_counts, mean_sentiments, on='App')

# Rename columns for clarity
final_df.rename(columns={
    'Positive': 'Positive_Sentiment',
    'Neutral': 'Neutral_Sentiment',
    'Negative': 'Negative_Sentiment',
    'Sentiment_Polarity': 'Mean_Sentiment_Polarity',
    'Sentiment_Subjectivity': 'Mean_Sentiment_Subjectivity'
}, inplace=True)

final_df.head()
````

````python
final_df.isna().sum()
````

````python
#merge two dataframe by inner join
merged_df=pd.merge(data,final_df,on='App',how='inner')
merged_df.head(4)
````

````python
# Chart - 6 visualization code

#create a group rating function
def app_rate(rating):
  '''
  This function helps to categorize the app rating
  '''
  if rating>=4.0:
    return 'Top Rated'
  elif rating>=3.0 and rating<4.0:
    return 'Above Average'
  elif rating>=2.0 and rating<3.0:
    return 'Average'
  else:
    return 'below average'
````
